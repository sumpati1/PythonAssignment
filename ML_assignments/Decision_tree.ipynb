```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris["DESCR"])
```

```python
print(iris["target"])
```

```python
#Independent feature
X = pd.DataFrame(iris["data"],columns=["sepal length","sepal width","petal length","petal width"])
```

```python
#Dependent feature
y = iris["target"]
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
#apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)  # pre pruning
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
#Visualize decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

Decision tree classifier pre-pruning and hyperparameter tuning

```python
param = {
    "criterion": ["gini","entropy","log_loss"],
    "splitter": ["best","random"],
    "max_depth": [1,2,3,4,5],
    "max_features": ["auto","sqrt","log2"],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treeclassifier,param_grid=param,cv=5,scoring="accuracy")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
score
```

Diabetes Prediction using Decision Tree Regressor

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset["DESCR"])
```

```python
df_diabetes = pd.DataFrame(dataset.data,columns= ['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
df_diabetes.head()
```

```python
X = df_diabetes
y = dataset["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 10)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot = True)
```

```python
 from sklearn.tree import DecisionTreeRegressor
 regressor = DecisionTreeRegressor()
 regressor.fit(X_train,y_train)
```

Hyperparameter Tuning

```python
param = {
    "criterion": ["absolute_error","friedman_mse","squared_error"],
    "splitter": ["best","random"],
    "max_depth": [1,2,3,4,5,10,15,20,25],
    "max_features": ["auto","sqrt","log2"],
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor,param_grid=param,cv=5,scoring="neg_mean_squared_error")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
selectmodel = DecisionTreeRegressor(criterion="friedman_mse",max_depth = 4,max_features = "log2",splitter="random")
```

```python
selectmodel.fit(X_train,y_train)
```

```python
#Visualize decision tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
a = tree.plot_tree(selectmodel,filled=True)
for txt in a:
    txt.set_color("black")
```
