```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#Import iris dataset
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
#Apply PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=iris.target)
```

```python
#Agglomerative Clustering
#To Construct a dendrogram
import scipy.cluster.hierarchy as sc
#Plot the Dendrogram
plt.figure(figsize=(20,7))
#plt.title("Dendrograms")

#Create Dendrogram
sc.dendrogram(sc.linkage(pca_scaled, method='ward'))  #ward tries to minimize variance within clusters.
plt.title("Dendrogram")                  #ward commonly uses Euclidean Distance.
plt.xlabel("Sample Index")
plt.ylabel("Euclidean Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering   #Only learns the data ,not transformation.
cluster = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```
