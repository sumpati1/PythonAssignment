```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#Load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
print(type(cancer_dataset)) # for bunch objects we use .keys() function
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
#Data preparation
df = pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
# Here 'data' is taken as independent variable(x) and 'feature_names' is taken as column names
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data  # means mean = 0 and SD = 1
```

```python
#Applying PCA algorithm
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_data  #Dimensionality reduction is done
```

```python
pca.explained_variance_   # variance of 2 features.  first PCA has maximum variance than second.
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```
